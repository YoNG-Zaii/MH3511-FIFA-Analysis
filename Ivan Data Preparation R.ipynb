#### Does the prestige level of the player’s team affect its wage or value? (2 sample t test)

Extract team with highest prestige vs team with lowest prestige and compare mean of both teams (players)

```r
playerPrestige <- subset(playersTeamCombined, select=c("Value.in.Euro.", "Wage.in.Euro.", "IntPrestige", "DomesticPrestige", "LeagueId", "League", "Club.Name"))
# filter to only intPrestige == 6 and intPrestige == 10
tier1And2Clubs <- playerPrestige[(playerPrestige$IntPrestige == 9) | (playerPrestige$IntPrestige == 10), ]
head(tier1And2Clubs); dim(tier1And2Clubs)
```

summary(tier1And2Clubs$IntPrestige)

```r
summary(tier1And2Clubs$IntPrestige)
```

```r
var.test(tier1And2Clubs$Value.in.Euro. ~ tier1And2Clubs$IntPrestige)
```

```r
t.test(tier1And2Clubs$Value.in.Euro. ~ tier1And2Clubs$IntPrestige, var.equal=T)
```

```r
var.test(tier1And2Clubs$Wage.in.Euro. ~ tier1And2Clubs$IntPrestige)
```

```r
t.test(tier1And2Clubs$Wage.in.Euro. ~ tier1And2Clubs$IntPrestige, var.equal=F)
```

Value and wage of a player depends on the prestige level of his club. From the results, a tier 1 club player are more valuable than that of a tier 2 club

What bout England Clubs since England Clubs are mostly filled with English players

```r
summary(playerPrestige$League)
```

```r
# get top 2 leagues from england england premier league (13), england league championship (14)
```

```r
onlyEnglishClubs <- playerPrestige[(playerPrestige$LeagueId == 13) | (playerPrestige$LeagueId == 14),  ]
head(onlyEnglishClubs)
```

```r
var.test(onlyEnglishClubs$Value.in.Euro. ~ onlyEnglishClubs$LeagueId)
```

```r
t.test(onlyEnglishClubs$Value.in.Euro. ~ onlyEnglishClubs$LeagueId, var.equal=F)
```

There is difference in value of both leagues (tier 1 vs tier 2)
